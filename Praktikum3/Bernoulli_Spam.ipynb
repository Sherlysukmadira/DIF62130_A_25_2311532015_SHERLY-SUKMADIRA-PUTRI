```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import os
for dirname, _, filenames in os.walk('SMSSpamCollection.csv'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

df = pd.read_table('SMSSpamCollection.csv',names=['label','sms_message'])

# Printing out first five columns
df.head()
```

**Data Preprocessing**

Kita perlu merubah label kita menjadi variabel biner, dimana 0 merepresentasikan 'ham/not spam' dan 1 merepresentasikan 'spam'

```python
df['label'] = df.label.map({'ham':0, 'spam':1})
```

```python
df.head()
```

```python
df.shape #ada 5572 baris dan 2 kolom
```

**Training dan Testing dataset**

dimana :

X_train training untuk 'sms_message' column.

y_train training untuk 'label' column

X_test testing data untuk 'sms_message' column.

y_test testing data untuk 'label' column.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['sms_message'],
                                                    df['label'],
                                                    random_state=1)

print('Number of rows in the total set: {}'.format(df.shape[0]))
print('Number of rows in the training set: {}'.format(X_train.shape[0]))
print('Number of rows in the test set: {}'.format(X_test.shape[0]))
```

```python
# Vectorizer binary=True untuk BernoulliNB (hanya 0 atau 1)
vectorizer = CountVectorizer(binary=True)

# Fit dan transform X_train
X_train_vector = vectorizer.fit_transform(X_train)

# Transform X_test (jangan fit ulang!)
X_test_vector = vectorizer.transform(X_test)
```

**Train Model Bernoulli Naive Bayes**

```python
# Inisialisasi model
bnb = BernoulliNB()

# Training
bnb.fit(X_train_vector, y_train)
```

**Evaluasi Model**

```python
# Prediksi data test
y_pred = bnb.predict(X_test_vector)

# Akurasi
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification Report (precision, recall, f1-score)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Contoh pesan baru
pesan_baru = ["Congratulations! You've won a free ticket to Bahamas. Call now!",
              "Meeting at 10 am, don't forget the report."]

# Vectorize pesan baru
pesan_baru_vector = vectorizer.transform(pesan_baru)

# Prediksi
prediksi = bnb.predict(pesan_baru_vector)

# Hasil
for i, pesan in enumerate(pesan_baru):
    label = "Spam" if prediksi[i] == 1 else "Ham"
    print(f"Pesan: {pesan} => Prediksi: {label}")
```
